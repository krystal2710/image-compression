## First part: Human Perception of Quality

### Load 20 images

```python
#SET UP

#import libraries
import pandas as pd
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
import k_means_lib
#set up path
import os
PATH = os.getcwd()
```

```python
data = []

#load 10 with-mask images
for i in range(10):
    file_link = PATH + '/dataset/with_mask/image_{}.png'.format(i)
    im = Image.open(file_link)
    data.append(np.asarray(im))

print('Done 10 with_mask images')

#load 10 without-mask images
for i in range(10):
    file_link = PATH + '/dataset/without_mask/image_{}.png'.format(i)
    im = Image.open(file_link)
    data.append(np.asarray(im))
print('Done 10 without_mask images')
```

```python
#take a look at 20 images

_, axs = plt.subplots(5, 4, figsize=(12, 12))
axs = axs.flatten()
for img, ax in zip(data, axs):
    ax.imshow(img)
plt.show()
```
